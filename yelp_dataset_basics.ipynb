```python
import json

def load_rows(filepath, nrows = None):
    with open(filepath) as json_file:
        count = 0
        objs = []
        line = json_file.readline()
        while (nrows is None or count < nrows) and line:
            count += 1
            obj = json.loads(line)
            objs.append(obj)
            line = json_file.readline()
        return pd.DataFrame(objs)
```

```python
business = load_rows('/kaggle/input/yelp_academic_dataset_business.json',10000)
print('Business objects loaded. Count = {}'.format(business.shape[0]))

tip = load_rows('/kaggle/input/yelp_academic_dataset_tip.json',10000)
print('tip objects loaded. Count = {}'.format(tip.shape[0]))

user = load_rows('/kaggle/input/yelp_academic_dataset_user.json',10000)
print('user objects loaded. Count = {}'.format(user.shape[0]))

review = load_rows('/kaggle/input/yelp_academic_dataset_review.json',10000)
print('checkin objects loaded. Count = {}'.format(review.shape[0]))

checkin = load_rows('/kaggle/input/yelp_academic_dataset_checkin.json',10000)
print('checkin objects loaded. Count = {}'.format(checkin.shape[0]))
```
